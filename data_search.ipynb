IQR for Baseline(RPM) Value

```python
variable = 'baseline value'

Q1 = df[variable].quantile(0.25)#Alt Çeyrek
Q3 = df[variable].quantile(0.75)#Üst Çeyrek
#Açıklık Hesabı
IQR = Q3 -Q1
#Sınırlar
lower_limit = Q1 - (1.5 * IQR)
upper_limit = Q3 + (1.5 * IQR)
print(f'{lower_limit} lower limit for {variable}')
print(f'{upper_limit} upper limit for {variable}')
#Outlier Filter
outliers = df[(df[variable] <lower_limit) | (df[variable] > upper_limit)]
print(f'Total number of outliers: {len(outliers)}')
print('Top five rows of outliers:')
print(outliers[variable].head())
```
